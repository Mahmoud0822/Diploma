```python
def clean_text(text):
  pattern = r"http(s)?:\/\/\S+"
  text = re.sub(pattern, '', text)
  pattern = r"\d+"
  text = re.sub(pattern, '', text)
  pattern = r"#\w*"
  text = re.sub(pattern, '', text)
  text = re.sub(r'^\s+', '', text)
  text= re.sub(r'\s+$','',text)
  text = re.sub(r'\s+'," ",text)
  text = re.sub(r'(.)\1{3,}',r'\1\1\1',text)
  return text
```

```python
def tokenize_text(text):
  tokens = text.split()
  return tokens
```

```python
def remove_stopwords(tokens):
  return [token for token in tokens if token not in stopwords_eng]
```

```python
stemmer = PorterStemmer()
def stemming(tokens):
  return [stemmer.stem(token) for token in tokens]
```

```python
def preprocess(text):
  text = clean_text(text)
  tokens = tokenize_text(text)
  tokens = remove_stopwords(tokens)
  tokens = stemming(tokens)
  return tokens
```

```python
tokens_positive = positive.apply(preprocess)
tokens_negative = negative.apply(preprocess)
tokens_neutral = neutral.apply(preprocess)
```

```python
def freq_table(tweets):
  freq = {}
  for tweet in tweets:
    for token in tweet:
      if token in freq:
        freq[token] += 1
      else:
        freq[token] = 1
  return freq
```

```python
freq_positive = freq_table(tokens_positive)
freq_negative = freq_table(tokens_negative)
freq_neutral = freq_table(tokens_neutral)
```

```python
def get_features(token):
  pos_sum=0
  neg_sum=0
  neutral_sum=0
  for w in token:
    pos_sum+=freq_positive.get(w,0)
    neg_sum+=freq_negative.get(w,0)
    # neutral_sum+=freq_neutral.get(w,0)
  return [pos_sum,neg_sum]
```

```python
x = []
y = []
for tweet in tokens_positive:
  x.append(get_features(tweet))
  y.append(1)
for tweet in tokens_negative:
  x.append(get_features(tweet))
  y.append(0)
# for tweet in tokens_neutral:
#   x.append(get_features(tweet))
#   y.append([0,0,1])
```

```python
combined = list(zip(x, y))
import random
random.shuffle(combined)
x, y = zip(*combined)
x = list(x)
y = list(y)
```
